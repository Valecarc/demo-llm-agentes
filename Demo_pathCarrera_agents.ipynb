```python
from crewai import Agent, Task, Crew, Process
from langchain_groq import ChatGroq
from langchain_community.tools import Tool

llm_llama=ChatGroq(temperature=0,
             model_name="llama-3.1-70b-versatile",
             api_key='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')
```

```python
def obtener_info_estudiante():
    carrera = input("¿Qué estás estudiando? ")
    intereses = input("¿Cuáles son tus principales intereses relacionados a lo que estas estudiando? (separa con comas) ")
    año = input("¿En qué año de estudio estás? ")
    experiencia = input("¿Tienes alguna experiencia laboral o en proyectos? Cuentame brevemente: ")

    return f"Carrera: {carrera}", f"Intereses: {intereses}", f"Año: {año}", f"Experiencia: {experiencia}"

```

```python
# Crear agentes con la herramienta de búsqueda
ana = Agent(
    role='Analista de Mercado laboral',
    goal="Investigar y proporcionar información lo mas actualizada posible, estratégica sobre las tendencias del mercado laboral relacionado a {carrera}"
         "que permita respaldar la toma de decisiones y la planificación de estrategias relacionadas con el mercado laboral en cuestion ",
    backstory="Analista de Mercado Senior con amplia experiencia en recopilación, análisis y presentación de datos relacionados con puestos de trabajo, "
              "demanda laboral y salarios de diversas industrias donde e involucra la profesion de {carrera} en Colombia."
              "Experto en identificar tendencias y patrones, brindando información valiosa para la toma de decisiones estratégicas. "
              "Excelentes habilidades analíticas y capacidad para interpretar datos complejos, utilizando herramientas avanzadas de análisis y modelado para ofrecesr sugerencias acertadas. "
              "Comprometida con la precisión y actualización constante de la información, asegurando la entrega de resultados confiables y relevantes. "
              "Fuerte capacidad de comunicación y presentación, facilitando la comprensión de los datos por parte del estudiante y integrantes del equipo de trabajo.",
    allow_delegation=False,
    llm=llm_llama
)

max = Agent(
    role='Mentor',
    goal="Porporcionar asesoria en base a la experiencia para crear un plan de desarrollo profesional personalizado y"
          " dar recomendaciones de recursos específicos para la profundizacion, especializacion, crecimiento y la búsqueda de empleo",
    backstory="Mentor Técnico y Especialista en temas de {carrera} , con amplia experiencia en evaluar el perfil de estudiantes y sugerir caminos de desarrollo profesional basados en las tendencias del mercado. "
      "Experto en la industria relacionada a  {carrera} con un profundo conocimiento de las habilidades, conocimientos y competencias requeridas para tener éxito en esta carrera, {intereses}. "
      "Capacidad demostrada para analizar las tendencias y demandas del mercado laboral y traducirlas en recomendaciones estratégicas para el crecimiento y desarrollo profesional de los estudiantes. "
      "Excelentes habilidades de comunicación y empatía con los estudiantes para guiarlos hacia el éxito en su carrera.",
    allow_delegation=False,
    llm=llm_llama
)

carlos = Agent(
    role='Consejero de Carrera',
    goal="Ayudar al estudiarnte a recopilar y organizar toda la informacion que ha conseguido para estructurar su plan de desarrollo profesional basados en las tendencias del mercado.",
    backstory="Consejero de Carrera con amplia experiencia en guiar a estudiantes y profesionales recién egresados en la construcción de un plan de carrera sólido y exitoso. "
              "Experto en brindar orientación y asesoramiento personalizado para ayudar a los individuos a identificar sus fortalezas, intereses y metas profesionales. "
              "Capacidad demostrada para desarrollar estrategias efectivas de búsqueda de empleo, mejorar habilidades de entrevista y establecer conexiones relevantes en la industria."
              "Excelentes habilidades de comunicación y empatía, facilitando una relación de confianza y apoyo con los estudiantes. "
              "Orientado a resultados y comprometido con el crecimiento y éxito profesional de los individuos.",
    allow_delegation=False,
    llm=llm_llama
)
```

```python
# Crear tareas
tarea_ana = Task(
    description=(
      "1. Recopilar informacion detallada sobre la situacion actual y las tendencias del mercado laboral para la carrera de {carrera}. "
      "2. incluye una lista de las 5 puestos más ofertados por las empresas a nivel nacional (Colombia), "
      "3. relacionando a cada uno de los puestos los skills o habilidades mas importantes y/o requeridos, "
      "4. recopila cual es el salario de entrada (para alguien que recien este iniciando) "
      " y cual puede ser el salario medio (alguien con experiencia) expresado en la moneda local (cop) para la lista de ofertas generada. "
      "5. presenta la informacion obtenida a modo de tabla: ofertas de trabajo, skills requeridos, salario de entrada y salario con experiencia"
      ),
    agent=ana,
    expected_output="Porporciona a los demas agentes la informacion obtenida sobre la situacion actual y tendencias del mercado,"
                    " puestos y salarios de manera detallada y clara para ser utilizada en las proximas tareas."
)

tarea_max = Task(
    description=(
        "1. Basándote en el perfil del estudiante: ({info_estudiante}) y en la información de mercado proporcionada en la tarea anterior"
        "indica cuales podrias ser los temas o áreas en las que el estudiante deberia profundizar y las habilidades clave a que deberia desarrollar "
        " para aspirar a cada una las ofertas mencionadas sugeridas y sugiere un path de estudio para cada una que le permita lograr un crecimiento profesional en esa area."
        "3. En cada una de las areas de profundizacion y habilidades clave a desarrollar recomienda recursos utiles para mas adelante crear un plan de estudios "
        " ya sea nombres de libros, cursos, seminarios, videos, certificaciones, etc, "
        ),
    agent=max,
    expected_output="Porporciona a los demas agentes la informacion obtenida de manera detallada sobre las opciones de paths de carrera/puestos de trabajo potenciales a la carrera mencionada, "
                    " cada una con sus respectivas áreas de profundización y habilidades clave para mas adelante crear un plan de accion."
)

tarea_carlos = Task(
    description=(
        "1. Recopila y utiliza toda la información proporcionada por los angentes en las tareas anteriores "
        "2. Crea un plan de desarrollo profesional en {carrera} para el estudiante ({info_estudiante}) basado en las ofertas y paths de carrera entregados en las tareas anteriores "
        "   - Plantea objetivos de estudio / plan de accion, a corto y largo plazo en relacion a las areas profundización y habilidades clave a desarrollar, "
        "   - Proporciona consejos y tips especificos y practicos enfocados a la carrera (no consejos genericos), para la búsqueda de empleo y preparación de entrevistas en el campo."
        "3. Organiza toda la informacion anterior de manera clara y legible en formato Markdown para que sea de facil entendimiento con la siguiente estructura:"
        "   - # Titulo inicial "
        "       - breve introduccion "
        "   (presenta las ofertas obtenidas por la analista Ana) "
        "   - # Situacion actual del mercado"
        "   - # Ofertas de trabajo relacionadas (tabla): (oferta|skills| salario base (cop)|salario promedio (cop))"
        "   (organiza y presenta la informacion generada por el mentor Max )"
        "   - # Algunos paths de carrera:"
        "       - ejemplos , areas de profundizacion y habilidades clave , recursos utiles"
        "   - # Plan de accion:"
        "     - Objetivos a corto plazo: "
        "     - Objetivos a largo plazo: "
        "   - # Recomendaciones para la búsqueda de empleo y preparación de entrevistas:"
        ),
    agent=carlos,
    expected_output="Un plan de desarrollo profesional completo y organizado en formato markdown "
                    "con todas la informacion reelevante proporcionada por lo agentes involucrados asi como consejos para la búsqueda de empleo y entrevistas."
)
```

```python
fun_estudiante = obtener_info_estudiante()
carrera=fun_estudiante[0]
intereses=fun_estudiante[1]
info_estudiante= '\n'.join(fun_estudiante)

# Crear y ejecutar el crew
crew = Crew(
    agents=[ana, max, carlos],
    tasks=[tarea_ana, tarea_max, tarea_carlos],
    verbose=True,
    process=Process.sequential
)

resultado = crew.kickoff(inputs={"carrera": carrera, "intereses": intereses, "info_estudiante": info_estudiante})

#print MD
from IPython.display import Markdown
Markdown(resultado)
```
